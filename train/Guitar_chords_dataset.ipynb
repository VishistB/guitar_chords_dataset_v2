```python
import cv2
import os
import uuid
```

```python
BASE_PATH =os.getcwd()
AM_PATH = os.path.join(BASE_PATH,'data', 'Am')
# BM_PATH = os.path.join(BASE_PATH,'data', 'Bm')
# C_PATH = os.path.join(BASE_PATH,'data', 'C')
D_PATH = os.path.join(BASE_PATH,'data', 'D')
# DM_PATH = os.path.join(BASE_PATH,'data', 'Dm')
F_PATH = os.path.join(BASE_PATH,'data', 'F')
G_PATH = os.path.join(BASE_PATH,'data', 'G')
```

```python
os.makedirs(AM_PATH)
# os.makedirs(BM_PATH)
# os.makedirs(C_PATH)
os.makedirs(D_PATH)
# os.makedirs(DM_PATH)
os.makedirs(F_PATH)
os.makedirs(G_PATH)
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened(): 
    ret, frame = cap.read()
    frame = frame[0:640,0:640, :]
    
    if cv2.waitKey(1) & 0XFF == ord('a'): 
        imgname = os.path.join(AM_PATH, '{}.jpg'.format(uuid.uuid1()))
        cv2.imwrite(imgname, frame)
    
    if cv2.waitKey(1) & 0XFF == ord('d'): 
        imgname = os.path.join(D_PATH, '{}.jpg'.format(uuid.uuid1()))
        cv2.imwrite(imgname, frame)
    if cv2.waitKey(1) & 0XFF == ord('f'):
        imgname = os.path.join(F_PATH, '{}.jpg'.format(uuid.uuid1()))
        cv2.imwrite(imgname, frame)
    if cv2.waitKey(1) & 0XFF == ord('g'):
        imgname = os.path.join(G_PATH, '{}.jpg'.format(uuid.uuid1()))
        cv2.imwrite(imgname, frame)
    
    
    cv2.imshow('Image Collection', frame)
    
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
